```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

# Step 1: Load real-world dataset
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv"
columns = ['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness',
           'Insulin', 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Outcome']
df = pd.read_csv(url, names=columns)

print("ðŸ“Š Dataset Preview:")
print(df.head())

# Step 2: Split features and labels
X = df.drop('Outcome', axis=1)
y = df['Outcome']

# Step 3: Preprocess (Normalize the features)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Step 4: Split into train and test
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

# Step 5: Build a Deep Learning model
model = Sequential([
    Dense(16, activation='relu', input_shape=(8,)),  # input layer (8 features)
    Dense(8, activation='relu'),                     # hidden layer
    Dense(1, activation='sigmoid')                   # output layer (0 or 1)
])

# Step 6: Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Step 7: Train the model
history = model.fit(X_train, y_train, epochs=100, batch_size=16, validation_split=0.1, verbose=0)

# Step 8: Evaluate the model
loss, accuracy = model.evaluate(X_test, y_test)
print(f"\nâœ… Model Accuracy on Real Data: {accuracy*100:.2f}%")

# Step 9: Make dynamic predictions
new_patient = np.array([[2, 140, 80, 25, 130, 35.0, 0.5, 29]])  # example input
new_patient_scaled = scaler.transform(new_patient)

prediction = model.predict(new_patient_scaled)
probability = prediction[0][0]

print(f"\nðŸ§ Predicted Diabetes Probability: {probability:.4f}")
print("ðŸ’Š Prediction:", "Has Diabetes" if probability > 0.5 else "No Diabetes")
```
